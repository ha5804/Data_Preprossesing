# CSV_DATA & MATPLOT

- DATA_READ

    (1): numpy 이용 하여 csv 데이터 읽어오기

        - np.genfromtxt("filename.csv", delimiter = ",")

    - data 파일에 저장된 point1.csv파일을 읽고 각 행을 x , y 로 구분해보자.

```python
filename = "./data/point1.csv"
#./는 현재 디렉토리를 의미한다. 모든 경로를 작성할 필요없이 같은 폴더내부 존재하면 위처럼 작성
```

- 경로를 저장했으면 csv파일의 데이터를 처리해보자.

    (csv 파일 내부는 , 를 기준으로 데이터가 나누어져 있다.)

    (csv 파일은 2행으로 나누어져 있다.)

```python
import numpy as np
import csv

point = np.genfromtxt(filename , delimiter = ',')
print(point.shape)
```

- point에 저장된 값이 2행 100열 임을 알 수 있다.

  각 행을 x , y로 표현해보자

```python
x = point[0]
y = point[1]
#2D 행렬기준 슬라이싱은 기본적으로 행을 선택한다.
```

- 처리한 csv 데이터를 plot 해보자.

  (1):x, y축의 글자 크기 수정

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 6))
#figure은 plot사이즈를 결정한다.
plt.plot(x, y, color = 'red', linestyle = ":")
plt.xticks(size = 8)
plt.yticks(size = 8)
plt.show()
```

### 3차원 데이터를 불러와서 plot까지 해보자!
(1) 3차원 데이터에 대해서는 새로운 import가 필요하다

    - from mpl_toolkits.mplot3d import Axes3D

```python
filename_3D = "./data/point2.csv"
point = np.genfromtxt(filename_3D, delimiter = ",")
x = point[0]
y = point[1]
z = point[2]
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
#2차원에서는 자동으로 fig,ax를 생성해주지만 3D에서는 직접 생성해야한다!
ax.scatter(x, y, z, s = 3, c = x , cmap = plt.cm.hsv_r)
#cmap을 활용하면 다양하게 색변화를 표현할수있다.
#이때 c인자를 통해 꼭 어떤 기준으로 색변화를 표현할건지 지정해야한다!
```

### 3차원데이터를 불러와 표면을 plot해보자!

(1) surface를 Plot할때는 각 1D배열을 2D배열로 grid해야한다.

    - ax.plot_surface


```python
from matplotlib import cm
#cm을 사용할수있게 하는 라이브러리
filename_3D_Surface = "./data/point3.csv"
point = np.genfromtxt(filename_3D_Surface, delimiter = ",")
x = point[0]
y = point[1]
z = point[2]
grid_x = x.reshape(100, 100)
grid_y = y.reshape(100, 100)
grid_z = z.reshape(100, 100)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(grid_x, grid_y, grid_z, cmap = cm.coolwarm)
```
